```python
import pandas as pd
import data
import graph
```

```python
def format_df(df, scale=None):
    if not scale is None:
        df = df.drop('scale', axis=1)
    if scale is None:
        scale=''
    df['name'] = df['name'].apply(lambda x: list(x.split('/'))[-1])
    df = df.drop('max_clique', axis=1)
    df = df.drop('is_right', axis=1)
    df = df.drop('wrong_vertices', axis=1)
    df = df.drop('Unnamed: 0', axis=1)
    df['time'] = df['time'].apply(lambda x: round(float(x), 2))
    df['time_'+str(scale)] = df['time']
    df['size_max_clique_'+str(scale)] = df['size_max_clique'].apply(lambda x : int(x)).astype(int)
    df = df.drop('time', axis=1)
    df = df.drop('size_max_clique', axis=1)
    df = df.drop('num_vertices', axis=1)
    return df
```

```python
pd.options.display.max_rows = 77
res = pd.read_csv('results/dimacs/DIMACS_results.csv')
res
```

```python
res['num_edges'] = res['name'].apply(lambda x: data.DataDIMACS.load('../DIMACS_all/'+x+'.clq').cnt_edges)
```

###### Данные отсортированы в порядке увеличения суммы количества вершин и ребер:

```python
res['sum_vertices_edges'] = res['num_vertices'] + res['num_edges']
res = res.sort_values(by='sum_vertices_edges')
res
```

```python
res.to_csv('dimacs_sum.csv', sep=',')
```

```python
res = res.dropna()
plt.style.use('ggplot')  
plt.rcParams['figure.figsize'] = (15, 5)
plt.xlabel('Сумма количества вершин и ребер')
plt.ylabel('Время работы')
plt.plot(res['sum_vertices_edges'], res['time_1.0'], label='time_1.0')
plt.plot(res['sum_vertices_edges'], res['time_0.1'], label='time_0.1')
plt.legend()
```

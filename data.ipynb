```python
import pandas as pd
```

```python
def format_df(df, scale=None):
    if not scale is None:
        df = df.drop('scale', axis=1)
    if scale is None:
        scale=''
    df['name'] = df['name'].apply(lambda x: list(x.split('/'))[-1])
    df = df.drop('max_clique', axis=1)
    df = df.drop('is_right', axis=1)
    df = df.drop('wrong_vertices', axis=1)
    df = df.drop('Unnamed: 0', axis=1)
    df['time'] = df['time'].apply(lambda x: round(float(x), 2))
    df['time_'+str(scale)] = df['time']
    df['size_max_clique_'+str(scale)] = df['size_max_clique'].apply(lambda x : int(x)).astype(int)
    df = df.drop('time', axis=1)
    df = df.drop('size_max_clique', axis=1)
    df = df.drop('num_vertices', axis=1)
    return df
```

```python
df1 = pd.read_csv('../results/cluster1/DIMACS_all_trust.csv', sep=';')
df2 = pd.read_csv('cluster2.csv', sep=';')
df1 = format_df(df1, 0.1)
df2 = format_df(df2, 1.0)
res = df1.merge(df2, 'left', on='name')
```

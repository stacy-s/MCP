```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def density(df):
    df['density'] = 2 * df['num_edges'] / (df['num_vertices'] * (df['num_vertices'] - 1))
    return df
```

```python
def build_density_time(df, name_col_x, name_col_y):
    plt.rcParams['figure.figsize'] = (15, 5)
    plt.xlabel(name_col_x)
    plt.ylabel(name_col_y)
    label = name_col_y
    plt.scatter(df[name_col_x], df[name_col_y], label=label)
    plt.legend()
```

## Результаты для набора DIMACS, отсортированные в порядке возрастания сумм количества вершин и ребер

```python
pd.options.display.max_rows = 77
dm = pd.read_csv('../results/dimacs_sum.csv', sep=',')
dm = dm.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis=1)
dm = density(dm)
dm = dm.dropna()

dm
```

## Разброс точек в зависимости плотности и времени при scale=0.1 на наборе даных DIMACS

```python
build_density_time(dm, name_col_x='density', name_col_y='time_0.1')
```

## Разброс точек в зависимости плотности и времени при scale=1.0 на наборе даных DIMACS

```python
build_density_time(dm, name_col_x='density', name_col_y='time_1.0')
```

## Результаты для набора графов Facebook

```python
pd.options.display.max_rows = 100
fb = pd.read_csv('../results/facebook/facebook_num_edges.csv', sep=';')
fb = density(fb)
fb
```

## Разброс точек в зависимости от плотности и времени при scale=1.0 на наборе даных Facebook

```python
build_density_time(fb, name_col_x='density', name_col_y='time_1.0')
```

## Разброс точек в зависимости от плотности и размера максимальной клики при scale=1.0 на наборе даных Facebook

```python
build_density_time(fb, name_col_x='density', name_col_y='size_max_clique_1.0')
```

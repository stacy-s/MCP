```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

##### Результаты для набора DIMACS, отсортированные в порядке возрастания сумм количества вершин и ребер

```python
pd.options.display.max_rows = 77
dm = pd.read_csv('dimacs_sum.csv', sep=',')
dm = dm.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis=1)
dm
```

##### График зависимости суммы количества вершин и ребер от времени

```python
dm = dm.dropna()
plt.rcParams['figure.figsize'] = (15, 5)
plt.xlabel('Сумма количества вершин и ребер')
plt.ylabel('Время работы')
plt.plot(dm['sum_vertices_edges'], dm['time_1.0'], label='time_1.0')
plt.plot(dm['sum_vertices_edges'], dm['time_0.1'], label='time_0.1')
plt.legend()
```
